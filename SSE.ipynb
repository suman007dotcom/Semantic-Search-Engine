```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
from sklearn.metrics.pairwise import cosine_similarity
```

```python
documents = [
    "Machine learning is a subset of artificial intelligence that allows systems to learn from data.",
    "Deep learning uses neural networks with many layers to model complex patterns in data.",
    "Gradient descent is an optimization algorithm used to minimize the loss function in machine learning.",
    "Neural networks are inspired by the structure of the human brain and consist of layers of neurons.",
    "Natural language processing enables computers to understand and generate human language.",
    "Support vector machines are supervised learning models used for classification and regression tasks.",
    "Backpropagation is the algorithm used to train neural networks by computing gradients.",
    "Overfitting occurs when a model learns the training data too well and fails to generalize.",
    "Python is the most popular programming language for machine learning and data science.",
    "Convolutional neural networks are widely used in image recognition and computer vision tasks."
    ]
```

```python
vec = TfidfVectorizer(stop_words = 'english')
vec
```

```python
X = vec.fit_transform(documents)   
print(X.shape)
```

```python
svd = TruncatedSVD(n_components = 4, random_state = 42)
X_lsa = svd.fit_transform(X)
print(X_lsa)
```

```python
#q = input("Query: ")
q = ["Explain what is Neural Networks"]
q_TfIdf = vec.transform(q)
```

```python
q_LA = svd.transform(q_TfIdf)
q_LA
```

```python
similarities = cosine_similarity(q_LA, X_lsa)[0]
similarities
```

```python
ranking = similarities.argsort()[::-1]
ranking
```

```python
for rank, i in enumerate(ranking):
    print(f"Rank {rank+1} (score={similarities[i]:.4f}): {documents[i]}")
```
